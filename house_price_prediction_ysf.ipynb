```python
# Her bir sayÄ±sal sÃ¼tun iÃ§in aykÄ±rÄ± deÄŸerleri gÃ¶rselleÅŸtirme ve tanÄ±mlama
numerical_columns = ['price', 'bedrooms', 'bathrooms', 'floors', 'waterfront', 'view', 'condition', 'grade']

for col in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[col])
    plt.title(col + " outliers")
    plt.show()

# AykÄ±rÄ± deÄŸerleri tanÄ±mlama
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outlier_step = 1.5 * IQR
    outlier_indices = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index
    return outlier_indices
```

```python
df_ysf = df[numerical_columns]
df_ysf.corr()
```

```python
# Calculate the correlation matrix
correlation_matrix = df_ysf.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation of Numerical Columns')
plt.show()
```

```python
# Box Plots
fig, axs = plt.subplots(8,1,dpi=145, figsize=(10,15))
i = 0
for col in df_ysf.columns:
    axs[i].boxplot(df[col], vert=False)
    axs[i].set_ylabel(col)
    i+=1
plt.show()
```

```python
# Create price bins 
df['price_bins'] = pd.qcut(df['price'], q=5)  # Divide the price by 5 according to counts

# Crate scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='lat', y='long', data=df, hue='price_bins', palette='coolwarm')
plt.show()
```

```python
# Box plots
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(x=df[column], y=df['price'])
    plt.title(f'{column.capitalize()} vs. Price')

plt.tight_layout()
plt.show()

# Violin plots
plt.figure(figsize=(15, 10))
for i, column in enumerate(numerical_columns):
    plt.subplot(3, 3, i + 1)
    sns.violinplot(x=df[column], y=df['price'])
    plt.title(f'{column.capitalize()} vs. Price')

plt.tight_layout()
plt.show()
```
